```python
# Make new columns  in the dataframe from existing data for the following attributes:
# City
# State
# ZipCode
# Private or Public
# for-profit or not-for-profit
# primarily associate's (1 for Yes, 0 for No)



##Split and add street address
df['address1'] = df['address'].str.split(',', n=1,expand=True)

##Add Address 2
address2 = df['address'].loc[df['address'].str.count(',')>2].str.split(',',n=1,expand=True)
address2 = address2[1].str.split(',',n=1,expand=True)
address2 = address2.drop(1,axis=1)
df['address2'] = address2

##Add City
city = df['address'].str.rsplit(',', n=2,expand=True)
city = city.drop(0,axis=1)
df['City'] = city[1]

##Add State
state = df['address'].str.rsplit(',', n=2,expand=True)
state = state[2].str.split(n=1, expand=True)
df['State'] = state[0]


##Add Zipcode
zipcode = df['address'].str.rsplit(' ', n=1,expand=True)
df['Zipcode'] = zipcode[1]
```

```python
##Add Private or Public column

df['Private or Public'] = df['Type'].apply(lambda x: "Public" in x)
df['Private or Public'] = np.where(df['Private or Public'], "Public", "Private")

##Add for-profit or not-for-profit column
df['for-profit or not-for-profit'] = df['Type'].apply(lambda x: 'not-for-profit' in x)
df['for-profit or not-for-profit'] = np.where(df['for-profit or not-for-profit'], 'not-for-profit', "for-profit")

##Add Primarily Associate's column
df["Primarily Associate's"] = df['Type'].apply(lambda x: "primarily associate's" in x)
df["Primarily Associate's"] = np.where(df["Primarily Associate's"], 1,0)


df
```

```python
## Using the new columns you've made above (and excluding the original type column),
## find both the mean and median size of each university by State, by Private or Public,
## and by whether the university is primarily associate's or not.

##Changed datatype for Primarily Associate's column
df["Primarily Associate's"] = df["Primarily Associate's"].astype(str)

##Mean and median by State
state_mean = df.groupby(['State']).mean().reset_index()
state_median = df.groupby(['State']).median().reset_index()

##Mean and median by Private or Public
private_public_mean = df.groupby(['Private or Public']).mean().reset_index()
private_public_median = df.groupby(['Private or Public']).median().reset_index()

df[['name', "Primarily Associate's"]]
```
