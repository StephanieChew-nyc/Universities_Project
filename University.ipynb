```python
# Make new columns  in the dataframe from existing data for the following attributes:
# City
# State
# ZipCode
# Private or Public
# for-profit or not-for-profit
# primarily associate's (1 for Yes, 0 for No)



##Split and add street address
df['address1'] = df['address'].str.split(',', n=1,expand=True)

##Add Address 2
address2 = df['address'].loc[df['address'].str.count(',')>2].str.split(',',n=1,expand=True)
address2 = address2[1].str.split(',',n=1,expand=True)
address2 = address2.drop(1,axis=1)
df['address2'] = address2

##Add City
city = df['address'].str.rsplit(',', n=2,expand=True)
city = city.drop(0,axis=1)
df['City'] = city[1]

##Add State
state = df['address'].str.rsplit(',', n=2,expand=True)
state = state[2].str.split(n=1, expand=True)
df['State'] = state[0]


##Add Zipcode
zipcode = df['address'].str.rsplit(' ', n=1,expand=True)
df['Zipcode'] = zipcode[1]
```
